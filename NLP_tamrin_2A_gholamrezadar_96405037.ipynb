### Imports

```
from matplotlib import pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from mpl_toolkits.mplot3d import Axes3D
```

### Loading Data

```

_path = "/content/drive/My Drive/Colab Notebooks/NLP Tamrins/ex2data1.txt"

train_x = []
train_y = []
test_x = []
test_y = []
training_perc = 70

with open(_path,"r") as data:
  raw_data = data.read().split("\n")[:-1]
  # traing data 
  train_x = [(float(i.split(",")[0]),float(i.split(",")[1])) for i in raw_data[:training_perc]]
  train_y = [float(i.split(",")[2]) for i in raw_data[:training_perc]]
  # test data
  test_x = [(float(i.split(",")[0]),float(i.split(",")[1])) for i in raw_data[training_perc:]]
  test_y = [float(i.split(",")[2]) for i in raw_data[training_perc:]]
```

### Logistic Regression

```

logistic_regression = LogisticRegression()

# Training
logistic_regression.fit(train_x,train_y)

# Prediction
y_pred = logistic_regression.predict(test_x)

# Confusion Matrix :   [true positive   ,   false positive]
#                      [false negative  ,    true negative]
a = confusion_matrix(test_y, y_pred)

# Accuracy
percentage = (a[0][0] + a[1][1]) / (len(test_x))
print("accuracy :", percentage*100,"%")

```

### Plotting

```
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
for i in range(len(train_x)):
    ax.scatter(train_x[i][0], train_x[i][1], train_y[i],c="g" if train_y[i]==0 else "b" )
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("class")
plt.show()
```
